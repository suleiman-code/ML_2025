# **Sentiment Analysis2 (multiple feedback at once time)**

```python
# Install the transformers library if not already installed
# !pip install transformers torch

from transformers import pipeline

def analyze_feedbacks(feedback_list):
    """
    Analyze sentiment for multiple feedbacks.
    feedback_list: list of strings
    Returns a list of dictionaries with text, sentiment, confidence, and emoji.
    """
    # Load the English sentiment analysis model
    classifier = pipeline(
        "sentiment-analysis",
        model="distilbert-base-uncased-finetuned-sst-2-english"
    )

    results = []
    for text in feedback_list:
        if not text.strip():
            continue  # Skip empty feedbacks
        res = classifier(text)[0]
        sentiment = res['label']
        confidence = res['score']

        # Assign emoji for visualization
        if sentiment.upper() == 'POSITIVE':
            emoji = "😊"
        elif sentiment.upper() == 'NEGATIVE':
            emoji = "😠"
        else:
            emoji = "😐"

        results.append({
            "text": text,
            "sentiment": sentiment,
            "confidence": confidence,
            "emoji": emoji
        })
    return results


def main():
    """
    Main function to run sentiment analysis for multiple user-entered feedbacks.
    """
    print("🚀 Loading the sentiment analysis model... Please wait.")

    try:
        # Take multiple feedbacks from the user
        feedbacks = []
        print("\n Enter your feedbacks one by one.")
        print(" Type 'done' when you finish entering feedbacks.\n")
        while True:
            user_input = input(" Enter feedback: ").strip()
            if user_input.lower() == 'done':
                break
            if user_input:
                feedbacks.append(user_input)
            else:
                print(" Empty feedback ignored.")

        if not feedbacks:
            print(" No feedbacks entered. Exiting program.")
            return

        print("\n Analyzing sentiments...\n")
        results = analyze_feedbacks(feedbacks)

        print(" Sentiment Analysis Results:")
        for res in results:
            print('******************************************************')
            print(f"{res['emoji']} Text: {res['text']}")
            print(f"Sentiment: {res['sentiment']}")
            print(f"Confidence: {res['confidence']:.4f}")
            print('******************************************************')

    except Exception as e:
        print(f" Error: {e}")


if __name__ == "__main__":
    main()
```

```python
from transformers import pipeline

# Load a sentiment-analysis pipeline (simpler and accurate than GPT-2 for this)
classifier = pipeline("sentiment-analysis")

# Take feedback from the user
user_feedback = input("👉 Enter your feedback: ").strip()

# Analyze sentiment
result = classifier(user_feedback)[0]

# Map sentiment to emoji
if result['label'].upper() == 'POSITIVE':
    emoji = "😊"
elif result['label'].upper() == 'NEGATIVE':
    emoji = "😠"
else:
    emoji = "😐"

# Print result
print(f"Sentiment Result: {result['label']} {emoji} (Confidence: {result['score']:.2f})")
```
