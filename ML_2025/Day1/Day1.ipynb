# **Data Preprocesssing part-1*

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ML_2025/Day1/Employee.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# display basic information
print('Data types of each Colum')
print(df.dtypes)
```

```python
 df.isnull().sum() # check missing values
```

```python
# handling missing values
# if age or city is missing ,fill with the middle values(median)
df['Age']=df['Age'].fillna(df['Age'].median())
df['PaymentTier']=df['PaymentTier'].fillna(df['PaymentTier'].median())
# fill with most accuring value
df['ExperienceInCurrentDomain']=df['ExperienceInCurrentDomain'].fillna(df['ExperienceInCurrentDomain'].mode()[0])
```

```python
# Label Encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder() # create LabelEncoder
df['Gender'] = le.fit_transform(df['Gender']) # Encode Gender column
df['Education'] = le.fit_transform(df['Education'])


```

```python
print(df.head()) # show first few rows
```

```python
# one-hot Encoding
# applying one hot encoding
encoded_df = pd.get_dummies(df['EverBenched'],prefix='EverBenched')
# concatenate with original dataset and drop old column
df = pd.concat([df, encoded_df,], axis=1).drop(columns='EverBenched')
```

```python
mms = MinMaxScaler()
df[['PaymentTier']] = mms.fit_transform(df[['PaymentTier']])

df['PaymentTier'].head()
```

# **Scaling**

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
df[['Age']] = sc.fit_transform(df[['Age']])
df[['Age']].head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#Label Encoding



le = LabelEncoder()

df['Education'] = le.fit_transform(df['Education'])
df['Education'].head()
```

```python
df.head()
```

```python
# One-Hot Encoding for 'Gender'
encoded_df = pd.get_dummies(df['Gender'], prefix='Gender',drop_first=True)

# Boolean ko 0 aur 1 me convert karo
encoded_df = encoded_df.astype(int)
# # Concatenate with original dataset and drop old 'Gender' column
df = pd.concat([df, encoded_df], axis=1).drop(columns='Gender')
```
